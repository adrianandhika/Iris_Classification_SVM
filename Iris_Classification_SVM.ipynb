```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()

from sklearn.svm import SVC
```

## Import Datasets


```python
from sklearn import datasets

iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
df = pd.DataFrame(np.c_[iris['data'],iris['target']], columns = iris['feature_names'] + ['target'])
df
```

```python
import matplotlib.pyplot as plt

sns.violinplot(y='target',x='sepal length (cm)',data=df,inner='quartile')
plt.show()
sns.violinplot(y='target',x='sepal width (cm)',data=df,inner='quartile')
plt.show()
sns.violinplot(y='target',x='petal length (cm)',data=df,inner='quartile')
plt.show()
sns.violinplot(y='target',x='petal width (cm)',data=df,inner='quartile')
plt.show()
```

```python
sns.pairplot(df, hue='target', markers='+')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap = 'cubehelix_r')
plt.show()
```

## Split the Model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

## Start the svm

```python
model = SVC(gamma='auto')
```

fit the model

```python
model.fit(X_train, y_train)
```

predict the model

```python
prediction = model.predict(X_test)
```

## Test the accuracy

```python
from sklearn.metrics import accuracy_score

print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
```
